```python
import pyttsx3
import speech_recognition as sr
from gtts import gTTS
import playsound
import os, sys, time
from datetime import datetime


engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)

class Chatbot:
    def __init__(self, disable_duration=1, default_key="OK GOOGLE", bye_key="BYE"):
        self.first_call = True
        self.disable_duration = disable_duration
        self.last_called = time.time()
        self.default_key = default_key
        self.bye_key = bye_key
        
    def speech(self, audio):
        engine.say(audio)
        engine.runAndWait()

    def takeCommand(self):
        r = sr.Recognizer()
        with sr.Microphone() as source:
            print("Listening...")
            r.adjust_for_ambient_noise(source)
            r.pause_threshold = 1
            r.energy_threshold = 1150
            audio = r.listen(source)
        try:
            print("recognize...")
            query = r.recognize_google(audio, language="bn-BD")
            print(f"User : {query}\n")
        except Exception as e:
            print(e)
            print("Say Again")
            return "None"
        return query
    
    def wakeup(self):
        w = sr.Recognizer()
        with sr.Microphone() as wakesource:
            print("Say 'Ok google' to run")
            w.adjust_for_ambient_noise(wakesource)
            w.pause_threshold = 1
            w.energy_threshold = 1150
            audio_wake = w.listen(wakesource)
        try:
            print("recognize...")
            wakeup_voice = w.recognize_google(audio_wake, language="en-EN").lower()
            print(f"User : {wakeup_voice.lower()}\n")
        except Exception as e:
            print(e)
            print("Say Again")
            return "None"
        return wakeup_voice 
     
    
    def wake_sleep(self):
        wakeup_voice = Chatbot().wakeup()
        if wakeup_voice == self.default_key.lower():
            Chatbot().speech("Hi! How can I help you?")
#             Chatbot().speech("হ্যালো, আপনাকে কিভাবে হেল্প করবো ?")
            Chatbot().takeCommand()
        elif wakeup_voice == self.bye_key.lower():
            Chatbot().speech("Ok, Take care")
        else:
            return None
```

```python
while True:
    p = Chatbot()
    r = p.wake_sleep()
#     s = p.takeCommand()
```

```python
import numpy as np
import cv2

cap = cv2.VideoCapture(0)

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Our operations on the frame come here
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    (h, w) = gray.shape[:2] #w:image-width and h:image-height
    img = cv2.circle(gray, (w//2, h//2), 7, (255, 255, 255), -1)
    img2 = cv2.rectangle(img, (160, 120),(480, 360), (255,0,0), 2)
    # Display the resulting frame
    crop_img = img[120:360, 160:480]
    cv2.imshow('frame',img)
    cv2.imshow("cutting", crop_img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
