```python
labels = []
question = []
answer = []
import string
from string import punctuation
for i in new_intents["intents"]:
    labels.append(i["tag"])
    for ques in i["patterns"]:
        question.append((ques, i["tag"]))
        if i["tag"] not in labels:
            labels.append(i["tag"])
            
#     for ans in i["responses"]:
#         answer.append((ans, i["tag"]))
#         if i["tag"] not in labels:
#             labels.append(i["tag"])
```

```python
myself = []
label_for_myself = []
for ques in question:
    myself.append(ques[0])
    label_for_myself.append(ques[1])
```

```python
label_for_myself
```

```python
max_sequence_length = max(myself, key=len)
len(max_sequence_length)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
tokenizer = Tokenizer()
tokenizer.fit_on_texts(myself)
```

```python
vocab = len(tokenizer.word_index)+1
```

```python
vocab_size = vocab
embedding_dim = 16
max_length = 40
trunc_type = "post"
oov_token = "<OOV>"
pad_type = "post"
```

```python
train_inputs = tokenizer.texts_to_sequences(myself)
train_inputs = pad_sequences(train_inputs, maxlen=40, padding=pad_type)
```

```python
train_inputs
```

```python
from keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
labelencoder_Y_1 = LabelEncoder()
yy = labelencoder_Y_1.fit_transform(label_for_myself)
yy = to_categorical(yy)
yy
```

```python
list(labelencoder_Y_1.classes_)
```

```python
labelencoder_Y_1.transform(labelencoder_Y_1.classes_)
```

```python
import tensorflow as tf
```

```python
input_shape = train_inputs.shape
```

```python
model =tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),
    tf.keras.layers.LSTM(200),
    tf.keras.layers.Flatten(), 
    tf.keras.layers.Dense(32, activation="relu"),
    tf.keras.layers.Dense(5, activation="softmax"),
])
model.compile(loss='categorical_crossentropy',optimizer="adam",metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(train_inputs, yy, epochs=500)
```
